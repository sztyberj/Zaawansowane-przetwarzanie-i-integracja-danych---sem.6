```python
#import libs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
#load data
data = pd.read_csv('https://raw.githubusercontent.com/sztyberj/Zaawansowane-przetwarzanie-i-integracja-danych---sem.6/main/Data/DanePogoda.csv',sep=';', encoding_errors="ignore")
data.head(10)
```

```python
#correct data
data.rename(columns={"zjawisko": "Zjawisko"}, inplace=True)
data.Zjawisko.unique()
```

```python
data.replace(["Mga"], "Mgła", inplace=True)
data.head(10)
```

```python
#Zadanie 1.1
for i in data.Godzina:
  if '06:00:00' <= i <= '17:00:00':
    data.Godzina.replace(i, 'rano', inplace=True)
  if '18:00:00' <= i <= '20:00:00':
    data.Godzina.replace(i, 'wieczór', inplace=True)
  if  i <= '05:00:00':
    data.Godzina.replace(i, 'noc', inplace=True)
  if i == '21:00:00' or i == '22:00:00' or i == '23:00:00':
    data.Godzina.replace(i, 'noc', inplace=True)
  
data.Godzina.head(20)
```

```python
#Zadanie 1.2
for i in data.Zachmurzenia:
  if i == 0:
    data.Zachmurzenia.replace(i, 'Brak', inplace=True)
  if 1 <= i <= 3:
    data.Zachmurzenia.replace(i, 'małe', inplace=True)
  if 4 <= i <= 5:
    data.Zachmurzenia.replace(i, 'umiarkowane', inplace=True)
  if 6 <= i <= 7:
    data.Zachmurzenia.replace(i, 'duże', inplace=True)
  if i == 8:
    data.Zachmurzenia.replace(i, 'pełne', inplace=True)

data.Zachmurzenia.head(20)
```

```python
#Zadanie 1.3
max = data.Cisnienie.max()
min = data.Cisnienie.min()
arr = []

for i in data.Cisnienie:
  norm_i = (i - min) / (max - min)
  arr.append(norm_i)

normalized_ser = pd.Series(arr)
data.Cisnienie.replace([data.Cisnienie], [normalized_ser], inplace=True)

data.Cisnienie.head(20)
```

```python
#Zadanie 4
data = pd.get_dummies(data, columns=['Godzina'])
data.head(20)
```
